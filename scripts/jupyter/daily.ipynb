# Kilowatt Hours (kWh) per Day
First look on daily data from Fronius Solarweb.
- Measerment for energy
- The kilowatt-hour is a composite unit of energy equal to one kilowatt (kW) of
power sustained for one hour.

## Prerequisits

```python
import pandas as pd
import matplotlib.pyplot as plt

pv = pd.read_csv('../../data/pv_day.csv', names=['Datum', 'kWh'], index_col='Datum', usecols=[0, 1], skiprows=[0, 1])
pv.index = pd.to_datetime(pv.index, format='%d.%m.%Y')
pv.index.freq = pd.infer_freq(pv.index)
pv.dtypes
```

## Time Series
- May to October

```python
pv.plot(grid=True, colormap='viridis', figsize=(14, 8), ylabel='kWh')
plt.show()
pv
```

## Distribution

```python
pv.plot(kind='box', figsize=(14,8), colormap='viridis').grid(axis='y')
plt.show()
pv.describe()
```

```python
pv.plot(kind='density', grid=True, figsize=(14, 8), colormap='viridis')
plt.show()
```
