```python
import pandas as pd
import matplotlib.pyplot as plt
```

## Einlesen der Daten
Date and Time
- Nachts: alle ca. 90 Minuten 4 Beobachtungen
- Tags: alle 30 Minuten eine Beobachtung

Inverter No.
- 1

Device Type
- 76

Periode [s]
- Abstand zwischen Beobachtungen in Sekunden
- Nachts NaN
- Tags 1800 (30 Minuten)

Energy positiv [Ws]
- Ws: Wattsekunden
- Nachts NaN
- Energie die in der letzten halben Stunde produziert wurde.

Reactive Energy [Vars]
- Nachts NaN
- Var: Voltampere Reaktiv
- 1 Var = 1 W
- Reactive Energy: Blindleistung
- tritt auf, wenn elektrische Energie über Wechselstrom transportiert wird
- lässt sich in induktive (L) und kapazitive (C) Last unterteilen

Uac L1 [V]
- V: Volt
- Spannung Wechselstrom
- Spannung an Ausgangsseite des Wechselrichters
- Spannung muss der Spannung im öffentlichen Stromnetz entsprechen

Iac L1 [A]
- A: Ampere
- anstehende Wechselstromstärke
- Leistungsfähigkeit der Anlage am Ausgang

Udc MPPT1[V] (Udc MPPT2[V]) 
- Wechselspannung am Eingang des Wechselrichters

Idc MPPT1[A] (Idc MPPT2[A])
- Strom am Eingang des Wechselrichters

Description
- Tags NaN
- Nachts: alle 90 Minuten 4 Einträge

```python
pv = pd.read_csv('../data/pv_30_min.csv', ';', skiprows=1)
pv.dtypes
```

## Datenaufbereitung
- Uac zwischen 230 und 240, außer einigen Ausreißern von 0
- Verlauf von Ws, Vars und A sieht änlich aus

```python
pv.index = pd.to_datetime(pv.Date.str.cat(pv.Time, ' '))
pv = pv.drop(columns=['Date', 'Time', 'Inverter No.', 'Device Type', 'Periode [s]', 'Description'])
pv = pv[pv.notna().any(1)]
pv
```

```python
fig, ax = plt.subplots(4, sharex=True, figsize=(14, 8))
pv['Energy positiv [Ws]'].plot(grid=True, colormap='viridis', ax=ax[0])
pv[['Reactive Energy L[Vars]', 'Reactive Energy C[Vars]']].plot(grid=True, colormap='viridis', ax=ax[1])
pv[['Uac L1 [V]', 'Udc MPPT1[V]', 'Udc MPPT2[V]']].plot(grid=True, colormap='viridis', ax=ax[2])
pv[['Iac L1 [A]', 'Idc MPPT1[A]', 'Idc MPPT2[A]']].plot(grid=True, colormap='viridis', ax=ax[3])
plt.show()
```

```python
fig, ax = plt.subplots(ncols=4, figsize=(14, 8))
pv['Energy positiv [Ws]'].plot(kind='box', grid=True, colormap='viridis', ax=ax[0])
pv[['Reactive Energy L[Vars]', 'Reactive Energy C[Vars]']].plot(kind='box', grid=True, colormap='viridis', 
                                                                ax=ax[1])
pv[['Uac L1 [V]', 'Udc MPPT1[V]', 'Udc MPPT2[V]']].plot(kind='box', grid=True, colormap='viridis', ax=ax[2])
pv[['Iac L1 [A]', 'Idc MPPT1[A]', 'Idc MPPT2[A]']].plot(kind='box', grid=True, colormap='viridis', ax=ax[3])
plt.show()
pv.describe()
```

```python
fig, ax = plt.subplots(ncols=4, figsize=(14, 8))
pv['Energy positiv [Ws]'].plot(kind='density', grid=True, colormap='viridis', ax=ax[0])
pv[['Reactive Energy L[Vars]', 'Reactive Energy C[Vars]']].plot(kind='density', grid=True, colormap='viridis', 
                                                                ax=ax[1])
pv[['Uac L1 [V]', 'Udc MPPT1[V]', 'Udc MPPT2[V]']].plot(kind='density', grid=True, colormap='viridis', ax=ax[2])
pv[['Iac L1 [A]', 'Idc MPPT1[A]', 'Idc MPPT2[A]']].plot(kind='density', grid=True, colormap='viridis', ax=ax[3])
plt.show()
```
